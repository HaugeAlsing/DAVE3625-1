Custom functions

```python
#this is unfinished, finish this function!

def grade_setter(x: int) -> str:
    
    if x == 0:
        return "Undefined Grade"
    if x < 18:
        return "F"
    elif 17 < x < 38:
        return "E"
    elif 37 < x < 58:
        return "D"
    elif 57 < x < 78:
        return "C"
    elif 77 < x < 93:
        return "B"
    elif x >= 94:
        return "A"
```
